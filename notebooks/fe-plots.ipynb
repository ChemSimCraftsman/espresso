```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
```

```python
dual4 = np.loadtxt("../src/fe/etot-vs-ecutwfc.dual4.dat")
dual8 = np.loadtxt("../src/fe/etot-vs-ecutwfc.dual8.dat")
dual12 = np.loadtxt("../src/fe/etot-vs-ecutwfc.dual12.dat")
```

```python
plt.plot(dual4[:, 0], dual4[:, 1], label="dual 4")
plt.plot(dual8[:, 0], dual8[:, 1], label="dual 8")
plt.plot(dual12[:, 0], dual12[:, 1], label="dual 12")
plt.legend(frameon=False)
plt.xlabel("ecutoff (Ry)")
plt.ylabel("Total energy (Ry)")
plt.xlim(dual4[0, 0], dual4[-1, 0])
plt.show()
```

```python
e, dos_up, dos_dn, _ = np.loadtxt('../src/fe/Fe.dos.dat', unpack=True)
```

```python
plt.plot(e, dos_up, linewidth=0.75, color='r', label='majority spin')
plt.plot(e, dos_dn*(-1), linewidth=0.75, color='g', label='minority spin')
plt.fill_between(e, 0, dos_up, where=(e < 12.843), facecolor='r', alpha=0.25)
plt.fill_between(e, 0, dos_dn*(-1), where=(e < 12.843), facecolor='g', alpha=0.25)
plt.axvline(x=12.843, linewidth=0.5, color='k', linestyle=(0, (5, 5)))
plt.text(12.2, -4.2, 'Fermi energy', fontsize=10, rotation=90)
plt.xlim(5, 20)
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.show()
```

```python
pdos_s = np.loadtxt('../src/fe/Fe.pdos.dat.pdos_atm#1(Fe)_wfc#1(s)')
pdos_d = np.loadtxt('../src/fe/Fe.pdos.dat.pdos_atm#1(Fe)_wfc#2(d)')
```

```python
plt.plot(pdos_s[:, 0], pdos_s[:, 1], lw=0.75, label="majority s-orbital")
plt.plot(pdos_s[:, 0], pdos_s[:, 2]*(-1), lw=0.75, label="minority s-orbital")
plt.plot(pdos_d[:, 0], pdos_d[:, 1], lw=0.75, label="majority d-orbital")
plt.plot(pdos_d[:, 0], pdos_d[:, 2]*(-1), lw=0.75, label="minority d-orbital")
plt.axvline(x=12.843, linewidth=0.5, color='k', linestyle=(0, (5, 5)))
plt.text(12.2, -4, 'Fermi energy', fontsize=10, rotation=90)
plt.xlim(5, 20)
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.legend(frameon=False)
plt.show()
```
