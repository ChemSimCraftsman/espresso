```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
```

```python
# Fermi energy is 15.7920 ev
# notice that in the earlier Co calculation we labelled col2 and col3 to majority
# and minority spins. Here, we plot atm#1 and atm#2 to majority and minority spins
# If we check the col2 and col3, they are just interchanged for Co atm 1, atm 2. 
pdos_co = np.loadtxt('../src/CoO/CoO.pdos_atm#1(Co1)_wfc#3(d)')
pdos_o = np.loadtxt('../src/CoO/CoO.pdos_atm#3(O)_wfc#2(p)')
```

```python
plt.plot(pdos_co[:, 0], pdos_co[:, 1], lw=0.75, label="Co-3d (majority)")
plt.plot(pdos_co[:, 0], pdos_co[:, 2], lw=0.75, label="Co-3d (minority)")
plt.plot(pdos_o[:, 0], pdos_o[:, 1], lw=0.75, label="O-2p")
plt.axvline(x=15.7920, linewidth=0.5, color='k', linestyle=(0, (5, 5)))
plt.text(15, 2.5, 'Fermi energy', fontsize=10, rotation=90)
plt.xlim(5, 20)
plt.ylim(0, )
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.legend(frameon=False)
plt.show()
```
