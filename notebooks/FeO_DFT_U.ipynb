```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
```

```python
# Fermi energy is 13.6466 ev
# notice that in the earlier Fe calculation we labelled col2 and col3 to majority
# and minority spins. Here, we plot atm#1 and atm#2 to majority and minority spins
# If we check the col2 and col3, they are just interchanged for Fe atm 1, atm 2. 
pdos_fe_majority = np.loadtxt('../src/FeO/feo_dos.dat.pdos_atm#1(Fe1)_wfc#5(d)')
pdos_fe_minority = np.loadtxt('../src/FeO/feo_dos.dat.pdos_atm#2(Fe2)_wfc#5(d)')
pdos_o = np.loadtxt('../src/FeO/feo_dos.dat.pdos_atm#3(O)_wfc#2(p)')
```

```python
plt.plot(pdos_fe_majority[:, 0], pdos_fe_majority[:, 1], lw=0.75, label="Fe-3d (majority)")
plt.plot(pdos_fe_minority[:, 0], pdos_fe_minority[:, 1], lw=0.75, label="Fe-3d (minority)")
plt.plot(pdos_o[:, 0], pdos_o[:, 1], lw=0.75, label="O-2p")
plt.axvline(x=13.6466, linewidth=0.5, color='k', linestyle=(0, (5, 5)))
plt.text(13, 4, 'Fermi energy', fontsize=10, rotation=90)
plt.xlim(6, 16)
plt.ylim(0, )
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.legend(frameon=False)
plt.show()
```
