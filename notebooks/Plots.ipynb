```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.figure(figsize = (8, 6))
plt.rcParams.update({'font.size': 16});
x, y = np.loadtxt('Etot-vs-ecutwfc.dat', delimiter=' ', unpack=True)
plt.plot(x, y, label='Etot vs ecutwfc')
plt.xlabel('ecutwfc')
plt.ylabel('Etot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.rcParams.update({'font.size': 16});
x, y = np.loadtxt('Etot-vs-kpoints.dat', delimiter=' ', unpack=True)
plt.plot(x, y, label='Etot vs kpoints')
plt.xlabel('kpoints')
plt.ylabel('Etot')
plt.legend()
plt.show()
```
