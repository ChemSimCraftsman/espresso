```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline

plt.rcParams.update(rcParamsDefault)
large = 22; med = 16; small = 14
params = {'legend.fontsize': med,
      'figure.figsize': (8, 6),
      'axes.labelsize': med,
      'axes.titlesize': med,
      'xtick.labelsize': med,
      'ytick.labelsize': med,
      'figure.titlesize': large,
      'image.origin': 'lower',
      'xtick.major.size': 4,
      'ytick.major.size': 4,
      'xtick.major.width': 0.5,
      'ytick.major.width': 0.5,
      'xtick.minor.size': 4,
      'xtick.minor.width': 0.5,
      'ytick.minor.size': 4,
      'ytick.minor.width': 0.5,
      'axes.linewidth': 0.75,
      'xtick.direction' : 'in', 
      'ytick.direction' : 'in', 
      'font.family': 'sans-serif', 
      'font.sans-serif' : 'Open Sans',
      'font.weight': 'light', # ultralight, light, regular, medium, bold 
      'font.style': 'normal', # normal, italic
}
plt.rcParams.update(params)

plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
```

```python
data = np.loadtxt('../src/Bi2Se3/bi2se3-bands-bulk-01.dat.gnu')

k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))

fig = plt.figure(figsize = (15, 6))
plt.subplot(131)
for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth=1, alpha=0.5, color='k')
plt.xlim(min(k), max(k))

# Fermi energy
plt.axhline(10.8017, linestyle=(0, (8, 10)), linewidth=0.75, color='k', alpha=0.5)
# High symmetry k-points (check bands_pp.out)
plt.axvline(0.5774, linewidth=0.75, color='k', alpha=0.5)
# text labels
plt.xticks(ticks= [0, 0.5774, k[-1]], labels=['M', '$\Gamma$', 'K'])
plt.ylabel("Energy (eV)")
plt.ylim(9.6, 12)
plt.title("Bulk")
plt.text(0.85, 10.85, 'Fermi energy', fontsize=12)

data = np.loadtxt('../src/Bi2Se3/bi2se3-bands-bulk-soc-02.dat.gnu')

k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))
plt.subplot(132)
for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth=1, alpha=0.5, color='k')
plt.xlim(min(k), max(k))
# Fermi energy
plt.axhline(10.5328, linestyle=(0, (8, 10)), linewidth=0.75, color='k', alpha=0.5)
# High symmetry k-points (check bands_pp.out)
plt.axvline(0.5774, linewidth=0.75, color='k', alpha=0.5)
# text labels
plt.xticks(ticks= [0, 0.5774, k[-1]], labels=['M', '$\Gamma$', 'K'])
plt.ylim(9.6, 12)
plt.title("Bulk+SOC")
plt.text(0.85, 10.6, 'Fermi energy', fontsize=12)

data = np.loadtxt('../src/Bi2Se3/bi2se3-bands-slab-soc-03.dat.gnu')

k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))
plt.subplot(133)
for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth=1, alpha=0.5, color='k')
plt.xlim(min(k), max(k))
# Fermi energy
plt.axhline(6.6483, linestyle=(0, (8, 10)), linewidth=0.75, color='k', alpha=0.5)
# High symmetry k-points (check bands_pp.out)
plt.axvline(0.5774, linewidth=0.75, color='k', alpha=0.5)
# text labels
plt.xticks(ticks= [0, 0.5774, k[-1]], labels=['M', '$\Gamma$', 'K'])
plt.ylim(5.6, 8)
plt.title("Surface+SOC")
# plt.text(0.85, 6.8, 'Fermi energy', fontsize=12)

plt.show()
```

```python
energy, dos, idos = np.loadtxt('../src/Bi2Se3/bi2se3_dos_04.dat', unpack=True)

fig = plt.figure(figsize = (8, 4))
plt.plot(energy, dos, linewidth=0.75, color='#006699')
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.axvline(x=6.65, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.xlim(6, 8)
plt.ylim(0, 30)
plt.fill_between(energy, 0, dos, where=(energy <= 6.65), facecolor='#006699', alpha=0.25)
plt.text(6.7, 20, 'Fermi energy', fontsize=12, rotation=90)
plt.show()
```
